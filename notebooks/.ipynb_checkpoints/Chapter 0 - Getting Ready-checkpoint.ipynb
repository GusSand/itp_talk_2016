# The Setup

This exercise has no code. It is simply the exercise you complete to get your computer to run Python, IPython Notebook and pandas.  Mac OS X computers already have Python 2, so do not install Python 3 (or any Python).

Do the following tasks:


### I know how to install python packages

You'll need an up-to-date version of IPython Notebook (&gt;= 3.0) and
pandas (&gt;=0.13) for this to work properly. It's set up to work with Python 2.7.

You can get these using pip:

```

    pip install ipython pandas numpy tornado pyzmq jinja2 matplotlib

```

### Create a workspace 

* Find your terminal program (use spotlight: command space) and put it in your dock. 
* create a directory to save your work for this class:

``

    md itpdata
    cd itpdata
``

### Run the python Interpreter

* In your terminal run python. You run things by just typing the name and hitting RETURN. 
* Type:

``

    print 'Hello world'

``

### Congrats you just created your first python program !

* Type quit(), Enter and get out of python
* You should be back at the prompt if not find out why. 


### Install an Editor and run it:

* Get a text editor (sublime, atom or textwrangler) and install it. 
* Launch your editor
* Put your editor in your dock. 
* Create a file that contains the following:

``

    print "Hello World!"
    print "Hello Again"
    print "I like typing this."
    
``

* Save the file as hello.py **in the directory you created on the previous step**
* Go back to the command line and type

``

    python hello.py

``
### Congrats you just created your second python program !
